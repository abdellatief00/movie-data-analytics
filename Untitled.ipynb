```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
df.dropna(axis = 0 ,how = 'any', inplace  = True)
df.isnull().sum()
```

```python
dir = df.groupby('director').score.agg(['mean' ,'min','max','count']).sort_values(by=['mean'] , ascending =[False ])
dir.loc[dir['count'] > 5].sort_values(by='mean' , ascending = False).head(10)
```

```python
df.loc[df['director'] == 'Martin Scorsese'].sort_values(by='score' , ascending = False)
```

```python
#number of movies per genre
plt.figure(figsize=(10,5))
count_genre = df.groupby('genre').genre.agg(['count']).sort_values('count', ascending =False)
sns.countplot(data  = df , x= 'genre').set(title = 'number of movies per genre')
plt.xticks(rotation  = 90)
plt.show()
```

```python
#gross per genre 
plt.figure(figsize= (18,7))
sns.catplot(kind = 'box' , data = df , x = 'genre', y = 'gross',whis=[5,95] ,aspect = 2 , height =4).set(title ='gross vs budget')
plt.xticks(rotation  = 90)
plt.show()
```

```python
#gross vs budget
us_gross = df[df['country'] == 'United States']
plt.figure(figsize=(18,7))
g = sns.regplot(
    data = us_gross , x = 'budget', y = 'gross' , scatter_kws = {'color' : 'c'} ,line_kws={'color':'red'} ).set(
    title='gross vs budget')
#sns.set_style('dark')
plt.show()
```

```python
#number of movies per year 
sns.displot(data = df , x= 'year' , kind = 'hist' ,aspect = 3 ,height = 5 ,discrete =True,color = 'red').set(
    title= 'number of movies per year ')
sns.set_style('dark')
plt.show()  
```

```python
#gross per year
plt.figure(figsize = (20,10))
sns.lineplot(data = df  , x = 'year',y = 'gross' ,hue='rating',ci = None ).set(title = 'gross per year')
plt.show()
```

```python
#coorelation heatmap between variables
corr_df = df.loc[:,['score','votes','budget','gross','runtime']].corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df ,annot  =True , vmin = -1 ,vmax = 1 , cmap = 'inferno' , linewidth = 1 ,linecolor = 'black' , center = 0 ,fmt  = '.1g')
heatmap.set_title('coorelation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
plt.show()
```

```python
#votes vs score
plt.figure(figsize=(6,6))
sns.jointplot(data = df, x= 'votes', y='score',color='red',kind='kde')
plt.title('votes vs score')
plt.show()
```

```python
#number of movies each year per country 
data_viz = df.groupby(['year','country'], as_index=False)['name'].count()
plt.figure(figsize=(20,10))
sns.lineplot(data = data_viz , x ='year', y = 'name',hue='country')
plt.show()
# for i in list(data_viz['transaction_type'].unique()):
#     viz = sns.lineplot(data=data_viz[data_viz['transaction_type'] == i], x='year', y='no_of_purchases', hue='tier')
#     plt.title(i)
#     plt.show()
```

```python
#runtime per genre 
plt.figure(figsize=(20,10))
sns.catplot(kind = 'boxen' , data = df , x = 'genre', y = 'runtime',aspect = 3 ,height = 4 ).set(title = "runtime per genre")
plt.xticks(rotation  = 90)
plt.show()
```

```python
#gross vs genre 
plt.figure(figsize=(20,10))
sns.stripplot(data =df , x= 'genre',y= 'gross',jitter = True).set(title = "gross vs genre")
plt.xticks(rotation =90)
plt.show()
```

```python
#median of votes per genre 
gg=df.groupby('genre' ,as_index =False).votes.median()
sns.catplot(kind = 'bar' , x= 'genre', y = 'votes', data =gg ,aspect = 3 , height = 3 ,palette ="viridis").set(title ='median of votes per genre')
plt.xticks(rotation = 90)
plt.show()
```

```python
#votes vs budget
plt.figure(figsize=(18,7))
g = sns.lmplot(
    data = df , x = 'votes', y = 'gross' , scatter_kws = {'color' : 'c'} ,line_kws={'color':'red'},aspect= 3 , height =4).set(
    title='votes vs budget' )
#sns.set_style('dark')
plt.show()
```

```python
#count of score
sns.displot(data =df , x = 'score' ,kind = 'hist' ,kde =True,aspect= 4 , height = 4,palette ='rocket' ,bins = 69).set(
title= 'count of score')
plt.show()
```

```python
#number of movies each year per genre
sns.displot(kind = 'hist', x ='year', hue ='genre', multiple ='stack' ,data = df,aspect= 4 , height = 6,discrete = True,
           palette = 'viridis').set(
title ='number of movies each year per genre')
plt.show()
```
